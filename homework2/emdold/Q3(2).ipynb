```python
import math
import numpy as np
import tensorflow as tf
```

```python
# sample n points from circle with radius r
def sample_from_circle(r, n):
    return [[math.cos(2 * math.pi / n * x) * r, math.sin(2 * math.pi / n * x) * r, 0] for x in range(0,n)]
```

```python
point_set = []
n = 60
m = 100
for r in np.arange(1.0, 10.0, 9 / m):
    pts = sample_from_circle(r, n)
    point_set.append(pts)
point_set = np.array(point_set)
```

```python
print(np.shape(point_set))
```

```python
d = tf.placeholder(tf.float32, [m, n, 3])

x = tf.get_variable('x',[1, n, 3], initializer=tf.random_normal_initializer(stddev=0.02))

X = tf.tile(x,[m,1,1])
```

```python
import tf_emddistance
dist,idx1,idx2=tf_emddistance.emd_distance(d,X)
L = tf.reduce_sum(dist)

learning_rate = 0.02
num_iter = 5000
optim = tf.train.AdamOptimizer(learning_rate).minimize(L)
```

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()

for i in range(num_iter):
    _, l = sess.run([optim,L], feed_dict={d:point_set})
    if i % 100 == 0:
        print(l)
```

```python
import matplotlib.pyplot as plt

wx = sess.run([x],feed_dict = {d: point_set})

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

for i in range(len(point_set)):
    plt.scatter(point_set[i][:, 0], point_set[i][:, 1], c="y")
plt.scatter([0], [0], c="y", label="original points")
plt.scatter(wx[0][0][:, 0], wx[0][0][:, 1], color='b', label="sampled points")
plt.axis('equal')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
